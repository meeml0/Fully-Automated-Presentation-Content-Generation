## Bolum 1: Kumeler ve Notasyon

**Temel hatirlatma:**
- Birlesim: $A \cup B$
- Kesisim: $A \cap B$
- Fark: $A \setminus B$
- Tumleyen: $A^c = U \setminus A$

Asagida evrensel kume ($U$) ve iki olay kumesi ($A$, $B$) uzerinde islemleri hesapliyoruz.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Rectangle

# Evrensel kume: bir ekipteki gelistiriciler
U = {"Ali", "Ayse", "Bora", "Ceren", "Deniz", "Ece", "Firat", "Gizem"}

# Olay kumeleri
A = {"Ali", "Ayse", "Deniz", "Gizem"}   # Python bilenler
B = {"Ayse", "Bora", "Ceren", "Deniz"}  # SQL bilenler

A_union_B = A | B
A_inter_B = A & B
A_diff_B = A - B
A_comp = U - A

print("U =", sorted(U))
print("A =", sorted(A))
print("B =", sorted(B))
print()
print("A U B =", sorted(A_union_B))
print("A n B =", sorted(A_inter_B))
print("A - B =", sorted(A_diff_B))
print("A^c   =", sorted(A_comp))
```

```python
# Gorsellestirme: Venn tipi kume semasi (ek paket gerektirmez)
fig, ax = plt.subplots(figsize=(8, 5))
ax.set_xlim(0, 10)
ax.set_ylim(0, 6)
ax.axis("off")

# Evrensel kume siniri
ax.add_patch(Rectangle((0.6, 0.6), 8.8, 4.8, fill=False, linewidth=2, edgecolor="black"))
ax.text(0.8, 5.1, "U", fontsize=12, fontweight="bold")

# A ve B kumeleri
ax.add_patch(Circle((4, 3), 1.8, color="#4C78A8", alpha=0.35))
ax.add_patch(Circle((6, 3), 1.8, color="#F58518", alpha=0.35))

ax.text(3.2, 3.0, "A-B", ha="center", va="center", fontsize=11)
ax.text(5.0, 3.0, "A n B", ha="center", va="center", fontsize=11, fontweight="bold")
ax.text(6.8, 3.0, "B-A", ha="center", va="center", fontsize=11)
ax.text(3.4, 4.8, "A", fontsize=12, color="#1f4e79", fontweight="bold")
ax.text(6.3, 4.8, "B", fontsize=12, color="#8a4b08", fontweight="bold")

ax.set_title("Kume Islemleri: Birlesim, Kesisim ve Fark")
plt.show()
```

## Bolum 2: Orneklem Uzayi ve Sistematik Listeleme

**Orneklem uzayi ($S$):** Bir deneyin tum olasi sonuclari.

Ayni deneyi farkli ayrinti seviyesinde temsil etmek mumkundur:
- **Detayli uzay ($S_1$):** tek tek sonuclar
- **Ozet uzay ($S_2$):** sonuclarin siniflandirilmis hali

```python
import itertools

# Deney: Iki urunun kalite kontrol sonucu (K: Kusurlu, S: Saglam)
states = ["K", "S"]
S1 = list(itertools.product(states, repeat=2))
print("S1 (detayli orneklem uzayi):", S1)
print("|S1| =", len(S1))

# Ayni deneyin ozet uzayi: kusurlu urun sayisi
S2 = [0, 1, 2]
print()
print("S2 (ozet orneklem uzayi):", S2)

# S1 -> S2 esleme
num_defective = {o: o.count("K") for o in S1}
print()
print("S1'den S2'ye esleme:")
for outcome, count in num_defective.items():
    print(f"{outcome} -> {count}")

# Olay: en fazla 1 kusurlu urun
A = [o for o in S1 if o.count("K") <= 1]
p_A = len(A) / len(S1)
print()
print(f"A = 'en fazla 1 kusurlu' olayi icin teorik P(A) = {p_A:.2f}")
```

```python
from collections import Counter

# Simulasyon ile dogrulama
np.random.seed(42)
N = 200000
sim_results = [(np.random.choice(states), np.random.choice(states)) for _ in range(N)]

sim_p_A = sum(1 for o in sim_results if o.count("K") <= 1) / N
print(f"Simulasyon P(A) ~= {sim_p_A:.4f}")
print(f"Teorik P(A) = {p_A:.4f}")

# Sonuc frekanslari
counter = Counter(sim_results)
labels = ["KK", "KS", "SK", "SS"]
counts = [
    counter[("K", "K")],
    counter[("K", "S")],
    counter[("S", "K")],
    counter[("S", "S")],
]

plt.figure(figsize=(7, 4))
plt.bar(labels, np.array(counts) / N, color=["#e45756", "#f58518", "#72b7b2", "#54a24b"])
plt.title("Simulasyonda Orneklem Uzayi Sonuc Dagilimi")
plt.ylabel("Goreli Frekans")
plt.xlabel("Sonuc")
plt.ylim(0, 0.30)
plt.show()
```

## Bolum 3: Agac Diyagrami ile Ardisik Deney

Agac diyagrami, **adim adim gerceklesen** deneylerde orneklem uzayini ve yol olasiliklarini gorsellestirir.

**Senaryo:**
1. Asama: Statik analiz sonucu `Temiz` (%70) veya `Uyari` (%30)
2. Asama:
   - `Temiz` ise manuel inceleme `Gecti` (%90), `Kaldi` (%10)
   - `Uyari` ise genis test `Gecti` (%40), `Kaldi` (%60)

```python
# Yol olasiliklari
p_temiz = 0.70
p_uyari = 0.30

p_gecti_temiz = 0.90
p_kaldi_temiz = 0.10

p_gecti_uyari = 0.40
p_kaldi_uyari = 0.60

paths = {
    ("Temiz", "Gecti"): p_temiz * p_gecti_temiz,
    ("Temiz", "Kaldi"): p_temiz * p_kaldi_temiz,
    ("Uyari", "Gecti"): p_uyari * p_gecti_uyari,
    ("Uyari", "Kaldi"): p_uyari * p_kaldi_uyari,
}

df_paths = pd.DataFrame(
    [{"Asama-1": k[0], "Asama-2": k[1], "Yol Olasiligi": v} for k, v in paths.items()]
)

print(df_paths)
print()
print(f"Toplam yol olasiligi = {df_paths['Yol Olasiligi'].sum():.2f}")

p_release = df_paths[df_paths["Asama-2"] == "Gecti"]["Yol Olasiligi"].sum()
print(f"P(Surum yayina cikar) = {p_release:.2f}")
```

```python
# Agac diyagramini ciz
fig, ax = plt.subplots(figsize=(11, 6))
ax.axis("off")

coord = {
    "Kok": (0.5, 0.5),
    "Temiz": (3.0, 0.75),
    "Uyari": (3.0, 0.25),
    "Temiz-Gecti": (6.2, 0.90),
    "Temiz-Kaldi": (6.2, 0.60),
    "Uyari-Gecti": (6.2, 0.40),
    "Uyari-Kaldi": (6.2, 0.10),
}

def draw_node(name, color="#e8eef7"):
    x, y = coord[name]
    ax.text(
        x, y, name, ha="center", va="center", fontsize=10,
        bbox=dict(boxstyle="round,pad=0.25", fc=color, ec="#2f4f6f", lw=1)
    )

def draw_edge(src, dst, label):
    x1, y1 = coord[src]
    x2, y2 = coord[dst]
    ax.annotate("", xy=(x2 - 0.35, y2), xytext=(x1 + 0.35, y1),
                arrowprops=dict(arrowstyle="->", lw=1.5, color="#2f4f6f"))
    ax.text((x1 + x2) / 2, (y1 + y2) / 2 + 0.04, label, fontsize=9, color="#2f4f6f")

for node in coord:
    color = "#fff2cc" if "Gecti" in node or "Kaldi" in node else "#e8eef7"
    draw_node(node, color)

draw_edge("Kok", "Temiz", "0.70")
draw_edge("Kok", "Uyari", "0.30")
draw_edge("Temiz", "Temiz-Gecti", "0.90")
draw_edge("Temiz", "Temiz-Kaldi", "0.10")
draw_edge("Uyari", "Uyari-Gecti", "0.40")
draw_edge("Uyari", "Uyari-Kaldi", "0.60")

ax.set_title("Agac Diyagrami: Yol Olasiliklari")
plt.show()
```

## Bolum 4: Olay Cebiri (Event Algebra)

Bir olay, orneklem uzayinin bir alt kumesidir.

Adil bir zar icin:
- $S = \{1,2,3,4,5,6\}$
- $A$: "cift gelmesi" = $\{2,4,6\}$
- $B$: "4 veya daha buyuk gelmesi" = $\{4,5,6\}$

Asagida $A \cup B$, $A \cap B$ ve tumleyen olaylari hesapliyoruz.

```python
# Olay kumeleri
S = set(range(1, 7))
A = {2, 4, 6}
B = {4, 5, 6}

A_union_B = A | B
A_inter_B = A & B
A_comp = S - A

p = lambda event: len(event) / len(S)

print("S =", S)
print("A =", A)
print("B =", B)
print()
print("A U B =", A_union_B, "=> P(A U B) =", p(A_union_B))
print("A n B =", A_inter_B, "=> P(A n B) =", p(A_inter_B))
print("A^c   =", A_comp, "=> P(A^c)   =", p(A_comp))
print()
print("Birlesim kurali kontrolu:")
print("P(A)+P(B)-P(A n B) =", p(A) + p(B) - p(A_inter_B))
print("A ve B ayrik mi?", A_inter_B == set())
```

```python
# Simulasyon ile birlesim kuralini dogrula
np.random.seed(7)
N = 300000
rolls = np.random.randint(1, 7, size=N)

sim_A = np.isin(rolls, list(A))
sim_B = np.isin(rolls, list(B))

sim_p_A = np.mean(sim_A)
sim_p_B = np.mean(sim_B)
sim_p_inter = np.mean(sim_A & sim_B)
sim_p_union = np.mean(sim_A | sim_B)

print(f"Simulasyon P(A)      ~= {sim_p_A:.4f}")
print(f"Simulasyon P(B)      ~= {sim_p_B:.4f}")
print(f"Simulasyon P(A n B)  ~= {sim_p_inter:.4f}")
print(f"Simulasyon P(A U B)  ~= {sim_p_union:.4f}")
print(f"Kuraldan P(A U B)    ~= {sim_p_A + sim_p_B - sim_p_inter:.4f}")

labels = ["A", "B", "A n B", "A U B"]
values = [sim_p_A, sim_p_B, sim_p_inter, sim_p_union]

plt.figure(figsize=(7, 4))
plt.bar(labels, values, color=["#4C78A8", "#F58518", "#72B7B2", "#54A24B"])
plt.ylim(0, 1)
plt.title("Olay Olasiliklari: Simulasyon Sonuclari")
plt.ylabel("Tahmini Olasilik")
plt.show()
```
