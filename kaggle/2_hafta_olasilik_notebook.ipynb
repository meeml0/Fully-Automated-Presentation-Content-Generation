## 📚 Bölüm 1: Koşullu Olasılık - Tablo Yöntemi

**Formül:** $P(A|B) = \frac{P(A \cap B)}{P(B)}$

**Senaryo:** Bir yazılım şirketindeki çalışanların teknoloji stack dağılımı.

| | Backend | Frontend | Toplam |
|---|---|---|---|
| **Python** | 40 | 10 | 50 |
| **JavaScript** | 15 | 35 | 50 |
| **Toplam** | 55 | 45 | 100 |

**Soru:** Bir çalışanın Backend geliştirici olduğu biliniyorsa, Python kullanma olasılığı nedir?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Tablo olarak veri
data = {
    'Backend': [40, 15],
    'Frontend': [10, 35]
}
df = pd.DataFrame(data, index=['Python', 'JavaScript'])
print(df)
print()

# Toplam çalışan
toplam = df.sum().sum()
print(f"Toplam Çalışan: {toplam}")

# P(Backend)
p_backend = df['Backend'].sum() / toplam
print(f"P(Backend) = {p_backend}")

# P(Python ∩ Backend)
p_python_ve_backend = df.loc['Python', 'Backend'] / toplam
print(f"P(Python ∩ Backend) = {p_python_ve_backend}")

# P(Python | Backend) - Koşullu Olasılık
p_python_given_backend = p_python_ve_backend / p_backend
print(f"\nP(Python | Backend) = {p_python_given_backend:.4f}")

# Direkt hesap (İndirgenmiş uzay)
direkt_hesap = df.loc['Python', 'Backend'] / df['Backend'].sum()
print(f"Direkt Hesap: {direkt_hesap:.4f}")
```

```python
# Görselleştirme: Heatmap
plt.figure(figsize=(8, 4))
sns.heatmap(df, annot=True, fmt='d', cmap='YlGnBu', cbar_kws={'label': 'Çalışan Sayısı'})
plt.title('Teknoloji Stack Dağılımı')
plt.ylabel('Programlama Dili')
plt.xlabel('Rol')
plt.show()
```

## 📚 Bölüm 2: Monte Carlo Simülasyonu - İadesiz Çekim

**Senaryo:** Bir torbada 5 kırmızı, 3 mavi top var. İki top art arda **iadesiz** çekiliyor.

**Soru:** İkinci topun mavi olma olasılığı nedir?

**Teorik:** Toplam olasılık teoremi ile $P(\text{Mavi}_2) = P(\text{Mavi}_2|\text{Kırmızı}_1)P(\text{Kırmızı}_1) + P(\text{Mavi}_2|\text{Mavi}_1)P(\text{Mavi}_1)$

```python
# Teorik Hesaplama
p_k1 = 5/8
p_m1 = 3/8

p_m2_given_k1 = 3/7  # 1. top kırmızı çıktıysa, torbada 5K, 3M kaldı (toplam 7)
p_m2_given_m1 = 2/7  # 1. top mavi çıktıysa, torbada 5K, 2M kaldı

p_m2 = (p_m2_given_k1 * p_k1) + (p_m2_given_m1 * p_m1)
print(f"Teorik P(Mavi₂) = {p_m2:.4f}")
```

```python
# Monte Carlo Simülasyonu
import random
random.seed(42)

def iadesiz_cekim_simulasyonu(deneme_sayisi=100000):
    torba = ['K']*5 + ['M']*3  # 5 Kırmızı, 3 Mavi
    mavi_2_sayac = 0
    
    for _ in range(deneme_sayisi):
        temp_torba = torba.copy()
        
        # İlk çekim
        top1 = random.choice(temp_torba)
        temp_torba.remove(top1)  # İadesiz
        
        # İkinci çekim
        top2 = random.choice(temp_torba)
        
        if top2 == 'M':
            mavi_2_sayac += 1
    
    return mavi_2_sayac / deneme_sayisi

sim_olasilik = iadesiz_cekim_simulasyonu()
print(f"Simülasyon P(Mavi₂) ≈ {sim_olasilik:.4f}")
print(f"Fark: {abs(p_m2 - sim_olasilik):.6f}")
```

## 📚 Bölüm 3: Bağımsız Olaylar

**Tanım:** İki olay $A$ ve $B$ bağımsızdır ancak ve ancak $P(A \cap B) = P(A) \cdot P(B)$

**Alternatif:** $P(A|B) = P(A)$ (B'nin gerçekleşmesi A'nın olasılığını değiştirmez)

**Senaryo:** İki sunucunun bağımsız olarak çalışırlığı (uptime).

```python
# Sunucu Güvenilirliği
p_sunucu_a = 0.98  # Sunucu A'nın çalışma olasılığı
p_sunucu_b = 0.95  # Sunucu B'nin çalışma olasılığı

# Bağımsızlık varsayımı altında
p_her_ikisi_de_calisir = p_sunucu_a * p_sunucu_b
print(f"P(Her iki sunucu da çalışır) = {p_her_ikisi_de_calisir:.4f}")

# En az biri çalışır (Sistem çalışır)
p_en_az_biri = 1 - ((1 - p_sunucu_a) * (1 - p_sunucu_b))
print(f"P(En az bir sunucu çalışır) = {p_en_az_biri:.4f}")

# Her ikisi de çöker (Sistem çöker)
p_sistem_coker = (1 - p_sunucu_a) * (1 - p_sunucu_b)
print(f"P(Sistem çöker) = {p_sistem_coker:.6f}")
```

```python
# Simülasyon
np.random.seed(42)
deneme = 1000000
sistem_calisir = 0

for _ in range(deneme):
    a_calisir = np.random.random() < p_sunucu_a
    b_calisir = np.random.random() < p_sunucu_b
    
    if a_calisir or b_calisir:  # En az biri çalışıyor
        sistem_calisir += 1

sim_p = sistem_calisir / deneme
print(f"\nSimülasyon: P(Sistem çalışır) ≈ {sim_p:.6f}")
print(f"Teorik: {p_en_az_biri:.6f}")
```

## 📚 Bölüm 4: Çarpım Kuralı ve Karar Ağaçları

**Çarpım Kuralı:** $P(A \cap B \cap C) = P(A) \cdot P(B|A) \cdot P(C|A \cap B)$

**Senaryo:** Yazılım testi süreci (3 aşama).

```python
# 3 aşamalı test süreci
p_unit_test_pass = 0.90
p_integration_pass_given_unit = 0.85
p_e2e_pass_given_prev = 0.80

# Tüm testleri geçme olasılığı
p_tum_testler_gecti = (p_unit_test_pass * 
                        p_integration_pass_given_unit * 
                        p_e2e_pass_given_prev)

print(f"P(Tüm testleri geçer) = {p_tum_testler_gecti:.4f}")
print(f"P(En az bir testte başarısız) = {1 - p_tum_testler_gecti:.4f}")
```

```python
# Görselleştirme: Karar Ağacı (Dallı Olasılık Gösterimi)
from matplotlib.patches import FancyBboxPatch, FancyArrowPatch

p_fail_unit = 1 - p_unit_test_pass
p_fail_int = 1 - p_integration_pass_given_unit
p_fail_e2e = 1 - p_e2e_pass_given_prev

fig, ax = plt.subplots(figsize=(13, 7))
ax.set_xlim(0, 12)
ax.set_ylim(0, 8)
ax.axis('off')

def ciz_dugum(x, y, metin, renk):
    kutu = FancyBboxPatch((x-0.9, y-0.32), 1.8, 0.64,
                          boxstyle='round,pad=0.2', edgecolor='black',
                          facecolor=renk, linewidth=1.2)
    ax.add_patch(kutu)
    ax.text(x, y, metin, ha='center', va='center', fontsize=9)

def ciz_ok(x1, y1, x2, y2, etiket, renk='black', dy=0.15):
    ok = FancyArrowPatch((x1, y1), (x2, y2), arrowstyle='->',
                         mutation_scale=12, linewidth=1.2, color=renk)
    ax.add_patch(ok)
    ax.text((x1+x2)/2, (y1+y2)/2 + dy, etiket, ha='center', fontsize=8, color=renk)

# Düğümler
ciz_dugum(6, 7, 'Başla', '#d8ecff')
ciz_dugum(3.5, 5.2, 'Unit Geç', '#c7f3d0')
ciz_dugum(8.5, 5.2, 'Unit Kal', '#ffd6d6')
ciz_dugum(2.2, 3.2, 'Integration Geç', '#baf0c2')
ciz_dugum(4.8, 3.2, 'Integration Kal', '#ffd6d6')
ciz_dugum(1.2, 1.2, 'E2E Geç', '#8fdf9d')
ciz_dugum(3.2, 1.2, 'E2E Kal', '#ffd6d6')

# Dallar ve olasılıklar
ciz_ok(5.3, 6.6, 4.0, 5.6, f'p={p_unit_test_pass:.2f}', 'green')
ciz_ok(6.7, 6.6, 8.0, 5.6, f'p={p_fail_unit:.2f}', 'red')
ciz_ok(3.2, 4.9, 2.4, 3.6, f'p={p_integration_pass_given_unit:.2f}', 'green')
ciz_ok(3.8, 4.9, 4.6, 3.6, f'p={p_fail_int:.2f}', 'red')
ciz_ok(1.9, 2.9, 1.4, 1.6, f'p={p_e2e_pass_given_prev:.2f}', 'green')
ciz_ok(2.5, 2.9, 3.0, 1.6, f'p={p_fail_e2e:.2f}', 'red')

# Yol olasılıkları
ax.text(0.2, 0.35, f'P(Tum testleri gecer) = {p_tum_testler_gecti:.4f}', fontsize=9, color='darkgreen')
ax.text(6.6, 4.45, f'P(Unitte kalma) = {p_fail_unit:.4f}', fontsize=9, color='darkred')
ax.text(5.1, 2.35, f'P(Integration kalma) = {p_unit_test_pass * p_fail_int:.4f}', fontsize=9, color='darkred')
ax.text(3.7, 0.35, f'P(E2E kalma) = {p_unit_test_pass * p_integration_pass_given_unit * p_fail_e2e:.4f}', fontsize=9, color='darkred')

plt.title('Test Süreci Karar Ağacı (Dallı Olasılık Gösterimi)', fontsize=13, weight='bold')
plt.tight_layout()
plt.show()
```
