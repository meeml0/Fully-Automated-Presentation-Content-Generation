## B?l?m 1: Toplam Olas?l?k Teoremi

Bir $S$ ?rnek uzay?, $\{B_1,\dots,B_n\}$ ile partition edilirse:

$$P(A)=\sum_{i=1}^{n} P(A\mid B_i)P(B_i)$$

Partition ko?ullar?:
- $B_i \cap B_j = \emptyset$ (i \neq j)
- $\bigcup_i B_i = S$
- $P(B_i)>0$

A?a??daki ?rnekte ?? farkl? ekibin toplam hataya katk?s?n? hesapl?yoruz.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ekip verileri (partition)
ekipler = ["Ekip A", "Ekip B", "Ekip C"]
p_ekip = [0.40, 0.35, 0.25]
p_hata_verilen_ekip = [0.02, 0.05, 0.08]

# Toplam olas?l?k
katkilar = [p_b * p_a_b for p_b, p_a_b in zip(p_ekip, p_hata_verilen_ekip)]
p_hata = sum(katkilar)

df_toplam = pd.DataFrame(
    {
        "Ekip": ekipler,
        "P(Ekip)": p_ekip,
        "P(Hata|Ekip)": p_hata_verilen_ekip,
        "P(Hata?Ekip)": katkilar,
    }
)

print("Toplam Olas?l?k Teoremi")
print(f"P(Hata) = {p_hata:.4f}")
print()
print(df_toplam)
print()
print(f"Toplam katk? kontrol?: {df_toplam['P(Hata?Ekip)'].sum():.4f}")
```

```python
# G?rselle?tirme: partition ve katk?lar
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 5))

ax1.pie(
    p_ekip,
    labels=ekipler,
    autopct="%1.1f%%",
    startangle=90,
    colors=sns.color_palette("pastel", len(ekipler)),
)
ax1.set_title("Partition: Ekip paylar?")

ax2.bar(df_toplam["Ekip"], df_toplam["P(Hata?Ekip)"], color=sns.color_palette("muted", len(ekipler)))
ax2.axhline(p_hata, color="red", linestyle="--", label=f"Toplam P(Hata) = {p_hata:.4f}")
ax2.set_ylabel("Katk? olas?l???")
ax2.set_title("Toplam hataya katk?lar")
ax2.legend()

plt.tight_layout()
plt.show()
```

## B?l?m 2: Ba??ms?zl?k ve Ayr?kl?k (Disjoint)

Temel ayr?m:
- **Ba??ms?zl?k:** $P(A\cap B)=P(A)P(B)$
- **Ayr?kl?k (disjoint):** $A\cap B=\emptyset$ ve dolay?s?yla $P(A\cap B)=0$

Not: Olas?l?klar? s?f?r olmayan olaylar i?in "ayr?k" olmak, "ba??ms?z" olmak anlam?na gelmez.


```python
# ?rnek 1 (ayr?k ama ba??ms?z de?il): tek zar
S_zar = set([1, 2, 3, 4, 5, 6])
D = {1}
E = {2}

def prob(olay, uzay):
    return len(olay) / len(uzay)

pD = prob(D, S_zar)
pE = prob(E, S_zar)
pDcapE = prob(D.intersection(E), S_zar)

# ?rnek 2 (ba??ms?z): iki para at???
S_para = {"YY", "YT", "TY", "TT"}
A = {"YY", "YT"}  # ilk at?? Y
B = {"YY", "TY"}  # ikinci at?? Y

pA = prob(A, S_para)
pB = prob(B, S_para)
pAcapB = prob(A.intersection(B), S_para)

sonuc = pd.DataFrame(
    [
        {
            "Durum": "Ayr?k ?rnek (D,E)",
            "P(X)": pD,
            "P(Y)": pE,
            "P(X?Y)": pDcapE,
            "P(X)P(Y)": pD * pE,
            "Ba??ms?z m??": np.isclose(pDcapE, pD * pE),
        },
        {
            "Durum": "Ba??ms?z ?rnek (A,B)",
            "P(X)": pA,
            "P(Y)": pB,
            "P(X?Y)": pAcapB,
            "P(X)P(Y)": pA * pB,
            "Ba??ms?z m??": np.isclose(pAcapB, pA * pB),
        },
    ]
)

print(sonuc)
```

```python
# G?rsel sezgi: ayr?k vs ba??ms?z
from matplotlib.patches import Circle

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Ayr?k
axes[0].add_patch(Circle((0.35, 0.5), 0.2, alpha=0.5, color="tab:blue"))
axes[0].add_patch(Circle((0.75, 0.5), 0.2, alpha=0.5, color="tab:orange"))
axes[0].set_title("Ayr?k olaylar (kesi?im yok)")
axes[0].text(0.25, 0.5, "D", fontsize=12)
axes[0].text(0.85, 0.5, "E", fontsize=12)
axes[0].set_xlim(0, 1.1)
axes[0].set_ylim(0, 1)
axes[0].axis("off")

# Ba??ms?z
axes[1].add_patch(Circle((0.45, 0.5), 0.25, alpha=0.5, color="tab:green"))
axes[1].add_patch(Circle((0.65, 0.5), 0.25, alpha=0.5, color="tab:red"))
axes[1].set_title("Ba??ms?z olaylar (kesi?im olabilir)")
axes[1].text(0.30, 0.5, "A", fontsize=12)
axes[1].text(0.80, 0.5, "B", fontsize=12)
axes[1].set_xlim(0, 1.1)
axes[1].set_ylim(0, 1)
axes[1].axis("off")

plt.tight_layout()
plt.show()
```

```python
# Ba??ms?zl?k uygulamas?: seri-paralel sistem
# C1 ve C2 paralel, bu blok C3 ile seri
p_c1, p_c2, p_c3 = 0.9, 0.9, 0.8

p_paralel = 1 - (1 - p_c1) * (1 - p_c2)
p_sistem = p_paralel * p_c3

print(f"P(C1 veya C2) = {p_paralel:.4f}")
print(f"P(Sistem) = P((C1?C2)?C3) = {p_sistem:.4f}")

# Basit Monte Carlo do?rulama
n = 200000
c1 = np.random.rand(n) < p_c1
c2 = np.random.rand(n) < p_c2
c3 = np.random.rand(n) < p_c3
sistem = (c1 | c2) & c3
print(f"Sim?lasyon ? {sistem.mean():.4f}")
```

## B?l?m 3: ?kili Ba??ms?zl?k vs Tam Ba??ms?zl?k

?? olay i?in tam ba??ms?zl?k kontrol?nde sadece ikili ko?ullar yetmez.

A?a??daki klasik kar??-?rnekte:
- $S=\{YY,YT,TY,TT\}$
- $A$: ilk at?? Y
- $B$: ikinci at?? Y
- $C$: iki at?? ayn?

Sonu?: Olaylar ikili ba??ms?zd?r ama tam ba??ms?z de?ildir.


```python
# Kar??-?rnek hesaplar?
S = {"YY", "YT", "TY", "TT"}
A = {"YY", "YT"}
B = {"YY", "TY"}
C = {"YY", "TT"}

def p(event):
    return len(event) / len(S)

pA, pB, pC = p(A), p(B), p(C)
pAB = p(A & B)
pAC = p(A & C)
pBC = p(B & C)
pABC = p(A & B & C)

print(f"P(A)={pA:.2f}, P(B)={pB:.2f}, P(C)={pC:.2f}")
print(f"P(A?B)={pAB:.2f}, P(A)P(B)={pA*pB:.2f}")
print(f"P(A?C)={pAC:.2f}, P(A)P(C)={pA*pC:.2f}")
print(f"P(B?C)={pBC:.2f}, P(B)P(C)={pB*pC:.2f}")
print(f"P(A?B?C)={pABC:.2f}, P(A)P(B)P(C)={pA*pB*pC:.2f}")

pairwise_independent = np.isclose(pAB, pA*pB) and np.isclose(pAC, pA*pC) and np.isclose(pBC, pB*pC)
mutually_independent = pairwise_independent and np.isclose(pABC, pA*pB*pC)

print()
print(f"?kili ba??ms?zl?k: {pairwise_independent}")
print(f"Tam ba??ms?zl?k: {mutually_independent}")
```

## B?l?m 4: Bayes Teoremi ve Sensitivite Analizi

Bayes form?l?:

$$P(H\mid +)=\frac{P(+\mid H)P(H)}{P(+\mid H)P(H)+P(+\mid H^c)P(H^c)}$$

Senaryo:
- $P(H)=0.001$
- $P(+\mid H)=0.99$
- $P(+\mid H^c)=0.05$

Ama?: Pozitif test sonras? ger?ek hasta olma olas?l???n? bulmak.


```python
# Bayes hesab? + sim?lasyon do?rulamas?
p_hasta = 0.001
p_saglikli = 1 - p_hasta
p_pozitif_hasta = 0.99
p_pozitif_saglikli = 0.05

p_pozitif = p_pozitif_hasta * p_hasta + p_pozitif_saglikli * p_saglikli
p_hasta_verilen_pozitif = (p_pozitif_hasta * p_hasta) / p_pozitif

print(f"P(Pozitif) = {p_pozitif:.6f}")
print(f"P(Hasta|Pozitif) = {p_hasta_verilen_pozitif:.4f} (%{100*p_hasta_verilen_pozitif:.2f})")

# Sim?lasyon
n = 1_000_000
hasta = np.random.rand(n) < p_hasta
pozitif = np.where(hasta, np.random.rand(n) < p_pozitif_hasta, np.random.rand(n) < p_pozitif_saglikli)

sim_deger = hasta[pozitif].mean()
print(f"Sim?lasyon do?rulamas? ? {sim_deger:.4f}")

cm = pd.DataFrame(
    {
        "Test Pozitif": [int((hasta & pozitif).sum()), int((~hasta & pozitif).sum())],
        "Test Negatif": [int((hasta & ~pozitif).sum()), int((~hasta & ~pozitif).sum())],
    },
    index=["Ger?ekte Hasta", "Ger?ekte Sa?l?kl?"],
)
print()
print("Confusion matrix")
print(cm)
```

```python
# Sensitivite analizi (3D yerine daha okunur heatmap)
sensitivite_aralik = np.linspace(0.90, 0.99, 10)
yanlis_pozitif_aralik = np.linspace(0.01, 0.10, 10)

kayitlar = []
for sens in sensitivite_aralik:
    for fp in yanlis_pozitif_aralik:
        p_pos = sens * p_hasta + fp * (1 - p_hasta)
        posterior = (sens * p_hasta) / p_pos
        kayitlar.append({"Sensitivite": sens, "Yanlis_Pozitif": fp, "P(Hasta|Pozitif)": posterior})

df_duyarlilik = pd.DataFrame(kayitlar)
pivot = df_duyarlilik.pivot(index="Yanlis_Pozitif", columns="Sensitivite", values="P(Hasta|Pozitif)")

plt.figure(figsize=(10, 6))
sns.heatmap(pivot, annot=True, fmt=".3f", cmap="YlOrRd")
plt.title("P(Hasta|Pozitif) i?in sensitivite analizi")
plt.xlabel("Sensitivite")
plt.ylabel("Yanl?? Pozitif Oran?")
plt.tight_layout()
plt.show()
```
